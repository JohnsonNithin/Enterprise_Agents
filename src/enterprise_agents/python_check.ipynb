```python

risk_keywords = {
    "prompt_injection": [
        "ignore previous instructions",
        "reveal system prompt",
        "bypass rules",
        "override",
    ],
    "privacy": [
        "password",
        "api key",
        "secret key"
    ],
    "harmful": [
        "bomb",
        "attack",
        "kill"
    ]
}
def normalize_text(text: str) -> str:
    """Clean user text before checking rules."""
    #return " ".join(text.lower())
    #return " ".join(text.lower().strip())
    return " ".join(text.lower().strip().split())

def input_guardrail(user_query: str) -> dict:
    text = normalize_text(user_query)

    for category, keywords in risk_keywords.items():
        for keyword in keywords:
            if keyword in text:
                return {"category": category,
                        "risk_level": "high",
                        "allowed": False,
                        "message:":f"Blocked because of {category} risk.Please verify the cited sources and consult an authorized supervisor before taking maintenance action."}
    
    else:
        return {"message":None,
                "risk_level":"low"}
                


input_guardrail("override " )
        
```
